# Retail Churn Prediction Model
This notebook performs EDA, feature engineering, model training (Logistic Regression, Random Forest, XGBoost), and evaluation. Theme: Green & White


## 1. Load data

```python
import pandas as pd

df = pd.read_csv('../data/retail_churn_data.csv')
df.head()
```

## 2. EDA & Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style='white')
plt.figure(figsize=(8,6))
plt.title('Churn distribution')
df['churn'].value_counts().plot(kind='bar')
plt.show()
```

## 3. Feature/Label split and basic model (Random Forest)

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score

features = ['tenure_months','avg_order_value','purchase_freq_per_month','support_tickets_last_6m','last_purchase_days_ago','total_orders','total_spent','is_vip']
X = df[features]
y = df['churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
y_prob = rf.predict_proba(X_test)[:,1]

print(classification_report(y_test, y_pred))
print('ROC AUC:', roc_auc_score(y_test, y_prob))
```

## 4. Save model and results


```python
import joblib
joblib.dump(rf, '../models/rf_churn_model.pkl')
```
